```python
import re

# Function to check password strength
def check_password_strength(password):
    strength = 0
    
    # Length
    if len(password) >= 8:
        strength += 1
    else:
        print("Password should be at least 8 characters long.")

    # Lowercase letters
    if re.search(r"[a-z]", password):
        strength += 1
    else:
        print("Add lowercase letters to make it stronger.")

    # Uppercase letters
    if re.search(r"[A-Z]", password):
        strength += 1
    else:
        print("Add uppercase letters to make it stronger.")

    # Numbers
    if re.search(r"[0-9]", password):
        strength += 1
    else:
        print("Add numbers to make it stronger.")

    # Special characters
    if re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):
        strength += 1
    else:
        print("Add special characters to make it stronger.")

    # Evaluate overall strength
    if strength == 5:
        print("Password Strength: Very Strong ✅")
    elif strength == 4:
        print("Password Strength: Strong 👍")
    elif strength == 3:
        print("Password Strength: Medium ⚠️")
    else:
        print("Password Strength: Weak ❌")

# User Input
password = input("Enter your password: ")
check_password_strength(password)

```
